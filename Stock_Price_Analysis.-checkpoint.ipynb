```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
from statsmodels.tsa.arima.model import ARIMA
from textblob import TextBlob
import requests
import datetime

# Set visualization style
plt.style.use('ggplot')

# Notebook configurations
pd.set_option('display.float_format', lambda x: f'{x:.2f}')
```

```python
# Function: Fetch Historical Stock Data
def fetch_stock_data(stock_ticker, start_date, end_date):
    print(f"Fetching data for {stock_ticker} from {start_date} to {end_date}")
    try:
        data = yf.download(stock_ticker, start=start_date, end=end_date)
        data['Return'] = data['Close'].pct_change()
        return data
    except Exception as e:
        print(f"Error fetching stock data: {e}")
        return None

# Function: Fetch News Articles
def fetch_news(stock_ticker):
    api_key = "YOUR_NEWS_API_KEY"  # Replace with your News API key
    url = f"https://newsapi.org/v2/everything?q={stock_ticker}&apiKey={api_key}"
    try:
        response = requests.get(url)
        articles = response.json().get("articles", [])
        return articles
    except Exception as e:
        print(f"Error fetching news: {e}")
        return []

# Function: Perform Sentiment Analysis
def analyze_sentiment(articles):
    sentiments = []
    for article in articles:
        title = article.get("title", "")
        analysis = TextBlob(title)
        sentiments.append(analysis.sentiment.polarity)  # Polarity: -1 to +1
    return sentiments

# Function: Calculate RSI
def calculate_rsi(data, window=14):
    delta = data['Close'].diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=window).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=window).mean()
    rs = gain / loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

# Function: Calculate MACD
def calculate_macd(data):
    data['EMA12'] = data['Close'].ewm(span=12, adjust=False).mean()
    data['EMA26'] = data['Close'].ewm(span=26, adjust=False).mean()
    data['MACD'] = data['EMA12'] - data['EMA26']
    data['Signal_Line'] = data['MACD'].ewm(span=9, adjust=False).mean()
    return data

# Function: Fetch Real-Time Stock Data
def fetch_live_data(stock_ticker):
    today = datetime.datetime.now().strftime("%Y-%m-%d")
    try:
        live_data = yf.download(stock_ticker, start=today, end=today)
        if live_data.empty:
            print("No live data available for today. Please try again during market hours.")
            return None
        return live_data
    except Exception as e:
        print(f"Error fetching live data: {e}")
        return None

# Function: Predict Next Day Price
def predict_next_day_price(model, live_data):
    if live_data is None or live_data.empty:
        print("Live data is unavailable, cannot make predictions.")
        return None

    try:
        last_price = live_data['Close'].iloc[-1]
        forecast = model.forecast(steps=1)[0]
        print(f"Last Close Price: {last_price}")
        print(f"Predicted Next Day Price: {forecast}")
        return forecast
    except Exception as e:
        print(f"Error making prediction: {e}")
        return None
```

```python
# Input Parameters
stock_ticker = "AAPL"  # Replace with desired stock ticker
start_date = "2015-01-01"
end_date = "2023-12-01"

# Step 1: Fetch Historical Data
data = fetch_stock_data(stock_ticker, start_date, end_date)

# Step 2: Visualize Historical Prices
if data is not None:
    plt.figure(figsize=(14, 7))
    plt.plot(data['Close'], label='Close Price')
    plt.title(f"{stock_ticker} Stock Price")
    plt.xlabel("Date")
    plt.ylabel("Close Price")
    plt.legend()
    plt.show()

# Step 3: Perform Sentiment Analysis
news_articles = fetch_news(stock_ticker)
if news_articles:
    sentiments = analyze_sentiment(news_articles)
    if sentiments:
        average_sentiment = sum(sentiments) / len(sentiments)
        print(f"Average Sentiment for {stock_ticker}: {average_sentiment:.2f}")

# Step 4: Add Technical Indicators
if data is not None:
    data['RSI'] = calculate_rsi(data)
    data = calculate_macd(data)

    # RSI Visualization
    plt.figure(figsize=(14, 7))
    plt.plot(data['RSI'], label='RSI', color='blue')
    plt.axhline(70, color='red', linestyle='--', label='Overbought (70)')
    plt.axhline(30, color='green', linestyle='--', label='Oversold (30)')
    plt.title('Relative Strength Index (RSI)')
    plt.xlabel('Date')
    plt.ylabel('RSI')
    plt.legend()
    plt.show()

    # MACD Visualization
    plt.figure(figsize=(14, 7))
    plt.plot(data['MACD'], label='MACD', color='blue')
    plt.plot(data['Signal_Line'], label='Signal Line', color='red')
    plt.axhline(0, color='black', linestyle='--', linewidth=0.8)
    plt.title('MACD (Moving Average Convergence Divergence)')
    plt.xlabel('Date')
    plt.ylabel('Value')
    plt.legend()
    plt.show()

# Step 5: Train ARIMA Model
if data is not None:
    model = ARIMA(data['Close'].dropna(), order=(5, 1, 0))  # Replace (p, d, q) with optimized values
    arima_result = model.fit()

    # Step 6: Real-Time Prediction
    live_data = fetch_live_data(stock_ticker)
    if live_data is not None:
        predicted_price = predict_next_day_price(arima_result, live_data)
```

```python
def fetch_stock_data(stock_ticker, start_date, end_date):
    print(f"Fetching data for {stock_ticker} from {start_date} to {end_date}")
    data = yf.download(stock_ticker, start=start_date, end=end_date)
    data.index = pd.to_datetime(data.index)  # Ensure the index is a DatetimeIndex
    data['Return'] = data['Close'].pct_change()
    return data
```

```python
def fetch_live_data(stock_ticker):
    today = datetime.datetime.now().strftime("%Y-%m-%d")
    live_data = yf.download(stock_ticker, start=today, end=today)
    if live_data.empty:
        print("No live data available for today. Please try again during market hours.")
        return None
    live_data.index = pd.to_datetime(live_data.index)  # Ensure a DatetimeIndex
    return live_data
```
